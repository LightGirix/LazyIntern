```python
pink_df = pd.read_csv("pink.csv")
metro_df = pd.read_csv("metro.csv")
```

```python
pink_df.head()
```

```python
metro_df.head()
```

```python
metro_df.columns = metro_df.columns.str.lower()
metro_df['line_code'] = metro_df['line_code'].astype(str)
metro_df['status'] = metro_df['status'].astype(bool)
```

```python
metro_df.head()
```

```python
merged = metro_df.merge(
    pink_df[['Pink/id', 'Pink/lat', 'Pink/long']],
    left_on='station_code',
    right_on='Pink/id',
    how='left',
)
merged[merged['line_code'] == 'Pink'].head()
```

```python
merged['latitude'] = merged['latitude'].combine_first(merged['Pink/lat'])
merged['longitude'] = merged['longitude'].combine_first(merged['Pink/long'])
merged[merged['line_code'] == 'Pink'].head()
```

```python
merged = merged.drop(columns=['Pink/id', 'Pink/lat', 'Pink/long'])
merged.head()
```

```python
metro_df = merged
metro_df[metro_df['line_code'] == 'Pink'].head()
```

```python
metro_df['status'] = metro_df['status'].mask(metro_df['line_code'] == 'Pink', True)
metro_df['status'] = metro_df['status'].mask(metro_df['line_code'] == 'Yellow', True)
metro_df[metro_df['line_code'] == 'Pink'].head()
```

```python
metro_df[metro_df['line_code'] == 'Yellow'].head()
```

```python
metro_df['line_code'] = metro_df['line_code'].str.lower()
metro_df.head()
```

```python
metro_df['line_code'] = metro_df['line_code'].str.replace(' ','_')
metro_df.head(100)
```

```python
metro_df = metro_df.drop(columns=["no."])
metro_df.head()
```

```python
metro_df = metro_df.rename(columns={'status': 'active'})
metro_df.head()
```

```python
metro_df.to_csv('new_metro.csv', index=False, encoding="utf-8-sig")
```
